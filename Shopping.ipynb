```
candy=['Red Jelly','Chocalate','Ice cream','Pink Cake','Mr. Donut','Brownies ','Jam Roll']
prize=[30,50,25,90,55,35,40]
bucket=[]
pocket=[]

def Menu():
    for i in range(0,6) :
        print(i+1,"\t",candy[i],"\t\t",prize[i])
        
def Check(money,pay):
    global c
    while money<pay :
        print("!!!Not enough money to pay!!!")
        c=Deal()
        if c==0:
            break
        money=Manage(money,c)
        

def Deal() :
    global bucket
    global pocket
    global money
    choice=int(input(">>>Pick your choice\n1.Delete this item\n2.Delete one on your bucket\n3.Leave the shop\nchoice : "))
    if choice==1 :
        return 0
    elif choice==2 :
        if len(bucket)==0:
            print("you have no item")
            return 0
        print("___bucket___")
        for i in range(0,len(bucket)):
            print(i+1,bucket[i],pocket[i])
        j=int(input("Delete : "))-1
        bucket.pop(j)
        c=pocket.pop(j)
        return c
    else :
        bucket=[]
        money=money+sum(pocket)
        pocket=[]
        return 0

def Manage(money,c):
    global shop
    if c==-1 :
        shop='n'
    money=money+c
    return money

print("~~~~~~ Welcome to CandyShop ~~~~~~")
shop='y'
c=1
money=int(input("Please input your money : "))
while(shop=='y'):
    print("\n\n")
    Menu()
    j=int(input("pick a number : "))-1
    pay=prize[j]
    Check(money,pay)
    if c!=0 :
        bucket.append(candy[j])
        pocket.append(prize[j])
        money=money-prize[j]
    shop=str(input(">>> Do you want to continue shopping? (y/n) : "))
print("\n\n_____________________________\n")
print("Here is your item : ")
for i in range(0,len(bucket)):
    print(i+1,bucket[i],pocket[i])
print("you have to pay",sum(pocket))
print("your change is", money)
print("-------------[END]-------------")
```
